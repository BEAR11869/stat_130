# 1

Simple Linear Regression: This statistical model aims to describe the linear relationship between two variables:
Predictor (X): The independent variable or feature.
Outcome (Y): The dependent variable or target, which we aim to predict or explain.
Model Equation:

The equation for simple linear regression is:
Y=β0+β1⋅X+ϵ

Error Term:
The error term ϵϵ is assumed to be normally distributed with a mean of zero. This means that any deviations from the predicted line are random and symmetrically distributed around the line.
    


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats



np.random.seed(42)  

X = np.linspace(0, 10, 100)


beta_0 = 2  # Intercept
beta_1 = 3  # Slope


epsilon = np.random.normal(loc=0, scale=1, size=len(X))


Y = beta_0 + beta_1 * X + epsilon

# Plot the data
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Sample Data')
plt.plot(X, beta_0 + beta_1 * X, color='red', label='True Line (without error)')
plt.xlabel('Predictor Variable (X)')
plt.ylabel('Outcome Variable (Y)')
plt.title('Simple Linear Regression: Predictor vs Outcome')
plt.legend()
plt.show()

# Fit a simple linear regression model using scipy
slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)

print(f"Estimated Intercept (beta_0): {intercept:.2f}")
print(f"Estimated Slope (beta_1): {slope:.2f}")
print(f"R-squared: {r_value**2:.2f}")
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf


np.random.seed(42)  
X = np.linspace(0, 10, 100)


beta_0 = 2  # Intercept
beta_1 = 3  # Slope


epsilon = np.random.normal(loc=0, scale=1, size=len(X))


Y = beta_0 + beta_1 * X + epsilon


data = pd.DataFrame({
    'X': X,
    'Y': Y
})

model = smf.ols(formula='Y ~ X', data=data).fit()

print(model.summary())

# Visualize the fitted regression line
plt.figure(figsize=(10, 6))
plt.scatter(data['X'], data['Y'], color='blue', label='Sample Data')
plt.plot(data['X'], model.fittedvalues, color='red', label='Fitted Line')
plt.xlabel('Predictor Variable (X)')
plt.ylabel('Outcome Variable (Y)')
plt.title('Fitted Simple Linear Regression Model')
plt.legend()
plt.show()
````

The fitted model's summary includes key tables, one of which shows the estimated parameters: the intercept (β0) and the slope (β1). These parameters are stored in the fitted_model.params attribute, which returns a series or array with the values:

fitted_model.params[0] or fitted_model.params.values[0]: The intercept (β0^).
fitted_model.params[1] or fitted_model.params.values[1]: The slope (β1^).

The fitted values (Y^Y^) are calculated using the equation:
Y^=β0^+β1^⋅X

This means that for each value of the predictor variable XX, the predicted YY value is computed by applying the estimated intercept and slope from the model.

So basically The fitted values give us the regression line's predicted Y values at each X point, forming a line that best fits the observed data according to the least squares criterion (minimizing the sum of squared residuals).
These predicted values are different from the true Y values due to the error term (ϵϵ) in the original data generation. The difference between the fitted line and the actual data points represents the residuals, capturing the randomness or unmodeled variability in the data.

In OLS regression, the chosen line (or fitted model) is the one that minimizes the total error between the observed values and the values predicted by the model. Specifically, the line is defined by the equation:
y=β0+β1x

The OLS method finds the values of β0 and β1β1​ that minimize the sum of the squared residuals, where a residual is the difference between the actual observed value yiyi​ and the predicted value y^i:
Residual=yi−y^i

The sum of squared residuals (SSR) is given by:
SSR=∑i=1n(yi−y^i)2


The reason for squares is Emphasis on Larger Errors, Squaring the residuals ensures that larger errors have a greater impact on the total sum. This helps the model prioritize reducing larger errors more than smaller ones, leading to a more balanced and effective fit. By squaring the residuals, all the terms become non-negative. This avoids the issue of positive and negative residuals canceling each other out, which would misrepresent the overall error. The squared terms make it possible to measure the total deviation from the line as a positive value, ensuring a meaningful minimization process.

|p-value|Evidence|
|-|-|
|$$p > 0.1$$|No evidence against the null hypothesis|
|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|

H0: There is no linear association between the independent variable xx (waiting time between eruptions) and the dependent variable YY (eruption duration). This can be formally stated as:

H0:β1=0


H1: There is a linear association between the independent variable and the dependent variable:

H1:β1≠0

If the p-value for β1 is < 0.05: This would provide strong evidence against the null hypothesis H0H0​, suggesting that there is a significant linear association between waiting and duration. The model suggests that waiting time is predictive of eruption duration.
If the p-value for β1 is ≥ 0.05: We fail to reject H0H0​, meaning that there isn't sufficient evidence to claim a linear association between waiting and duration.

````python
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

# Filter data for long wait times
long_wait_limit = 71
long_wait_data = old_faithful[old_faithful.waiting > long_wait_limit]

# Print original model summary for comparison
print(smf.ols('duration ~ waiting', data=long_wait_data).fit().summary().tables[1])

# Bootstrap slope coefficients
bootstrap_slopes = [
    smf.ols('duration ~ waiting', data=long_wait_data.sample(n=len(long_wait_data), replace=True)).fit().params['waiting']
    for _ in range(1000)
]

# Visualize bootstrapped slope distribution and 95% CI
plt.hist(bootstrap_slopes, bins=30, edgecolor='k')
plt.title('Bootstrapped Slope Coefficients')
plt.xlabel('Slope')
plt.ylabel('Frequency')
plt.show()

ci_lower, ci_upper = np.percentile(bootstrap_slopes, [2.5, 97.5])
print(f"95% CI for slope: [{ci_lower:.4f}, {ci_upper:.4f}]")
print(f"Contains 0? {'Yes' if ci_lower <= 0 <= ci_upper else 'No'}")

# Simulate null hypothesis data and fit models
np.random.seed(42)
simulated_slopes = [
    smf.ols('duration ~ waiting', data=long_wait_data.assign(duration=1.65 + np.random.normal(0, 0.37, len(long_wait_data)))).fit().params['waiting']
    for _ in range(1000)
]

# Visualize simulated slope distribution under H0
plt.hist(simulated_slopes, bins=30, edgecolor='k', color='orange')
plt.title('Simulated Slopes Under H0')
plt.xlabel('Slope')
plt.ylabel('Frequency')
plt.show()

# Compare p-values
original_model = smf.ols('duration ~ waiting', data=long_wait_data).fit()
simulated_p_value = (np.abs(simulated_slopes) >= np.abs(original_model.params['waiting'])).mean()

print(f"Original p-value: {original_model.pvalues['waiting']:.4e}")
print(f"Simulated p-value: {simulated_p_value:.4e}")
````

### 13. The "short" and "long" wait times are not "before and after" measurements so there are not natural pairs on which to base differences on which to do a "one sample" (paired differences) *hypothesis test*; but, we can do "two sample" hypothesis testing using a *permuation test*, or create a 95% *bootstrap confidence interval* for the difference in means of the two populations. 

### (A) Do a permuation test $\;H_0: \mu_{\text{short}}=\mu_{\text{long}} \; \text{ no difference in duration between short and long groups}$ by "shuffling" the labels
### (B) Create a 95% bootstrap confidence interval  by repeatedly bootstrapping within each group and applying *np.quantile(bootstrapped_mean_differences, [0.025, 0.975])* to the collection of differences between the sample means.    
### (a) Explain how the sampling approaches work for the two simulations.
### (b) Compare and contrast these two methods with the *indicator variable* based model approach used in Question 11, explaining how they're similar and different.<br>
    
<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>

> _You'll need to create `for` loops for repeated (shuffling simulation) **permutation** and (subgroup) **bootstrapping**, where_
>
> - _"shuffling" for **permutation testing** is done like this `old_faithful.assign(kind_shuffled=old_faithful['kind'].sample(n=len(old_faithful), replace=False).values)#.groupby('kind').size()`; then, the **mean difference statistic** is then calculated using `.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]` (so the **observed statistic** is `old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]`_
> 
>
> - _"two sample" **bootstrapping** is done like this `old_faithful.groupby('kind').apply(lambda x: x.sample(n=len(x), replace=True)).reset_index(drop=True)#.groupby('kind').size()`; then, the **bootstrapped mean difference statistic** is then calculated using `.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]` (like the **observed statistic** except this is applied to the **bootstrapped** resampling of `old_faithful`)_

````python
#A

np.random.seed(42)
observed_mean_diff = old_faithful[old_faithful.kind == 'long']['duration'].mean() - \
                     old_faithful[old_faithful.kind == 'short']['duration'].mean()

n_permutations = 10000
mean_diffs = []

for _ in range(n_permutations):
    shuffled_labels = np.random.permutation(old_faithful['kind'])
    group1_mean = old_faithful[shuffled_labels == 'short']['duration'].mean()
    group2_mean = old_faithful[shuffled_labels == 'long']['duration'].mean()
    mean_diffs.append(group2_mean - group1_mean)

p_value = (np.abs(mean_diffs) >= np.abs(observed_mean_diff)).mean()
print(f"Permutation Test p-value: {p_value:.4f}")
````

````python
#B 
np.random.seed(42)
n_bootstrap = 10000
boot_mean_diffs = []

for _ in range(n_bootstrap):
    short_sample = old_faithful[old_faithful.kind == 'short']['duration'].sample(frac=1, replace=True).mean()
    long_sample = old_faithful[old_faithful.kind == 'long']['duration'].sample(frac=1, replace=True).mean()
    boot_mean_diffs.append(long_sample - short_sample)

ci_lower, ci_upper = np.quantile(boot_mean_diffs, [0.025, 0.975])
print(f"95% Bootstrap Confidence Interval: [{ci_lower:.4f}, {ci_upper:.4f}]")
````

# a

Permutation Test:Tests the null hypothesis by assuming no difference between the groups and generating a distribution by random label shuffling.The p-value is derived by checking how often the observed mean difference falls in the extreme of this generated distribution.

Bootstrap Method:Estimates the sampling distribution of the difference in means by resampling from the original groups with replacement.The 95% confidence interval represents the range of plausible values for the true difference in means between the groups.

# b

Similarities:Both methods assess differences between "short" and "long" groups.They provide a way to infer whether the observed difference in means is statistically significant.The indicator variable model (βcontrast) essentially tests whether there is a significant mean difference between the categories, akin to testing μshort≠μlong

Differences:Permutation Test and Bootstrap:These are non-parametric approaches and do not rely on assumptions about the underlying data distribution.They directly compare observed data by shuffling or resampling and do not require model fitting.

The **p-values** provided by `statsmodels` via `smf.ols(...).fit()` depend on these assumptions, so if they are not (at least approximately) correct, the **p-values** (and any subsequent claims regarding the "evidience against" the **null hypothesis**) are not reliable. In light of this consideration, describe how you could diagnostically check the first three assumptions (given below) when using analyses based on **Simple Linear regression** model. From an Ethical and Professional perspective, do you think doing diagnostic checks on the assumptions of a **Simple Linear regression** model is something you can and should do whenever you're doing this kind of analysis? 
            
> The first three assumptions associated with the **Simple Linear regression** model are that
> 
> - the $\epsilon_i$ **errors** (sometimes referred to as the **noise**) are **normally distributed**
> - the $\epsilon_i$ **errors** are **homoscedastic** (so their distributional variance $\sigma^2$ does not change as a function of $x_i$)
> - the linear form is [at least reasonably approximately] "true" (in the sense that the above two remain [at least reasonably approximately] "true") so that then behavior of the $Y_i$ **outcomes** are represented/determined on average by the **linear equation**)<br>
> 
>    and there are additional assumptions; but, a deeper reflection on these is "beyond the scope" of STA130; nonetheless, they are that<br><br>
> - the $x_i$ **predictor variable** is **measured without error**
> - and the $\epsilon_i$ **errors** are **statistically independent** (so their values do not depend on each other)
> - and the $\epsilon_i$ **errors** are **unbiased** relative to the **expected value** of **outcome** $E[Y_i|x_i]=\beta_0 + \beta_1x_i$ (which is equivalently stated by saying that the mean of the **error distribution** is $0$, or again equivalently, that the **expected value** of the **errors** $E[\epsilon_i] = 0$)
